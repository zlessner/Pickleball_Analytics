```python
import random
```

```python
gameEnd = 3

probAServe = .5

probBServe = .5

probA = probAServe


iterations = 100000


teamAScore = 0

teamBScore = 0

winby2 = abs(teamAScore - teamBScore)

servingTeam = 1

teamAWins = 0

teamBWins = 0


```

```python
#Singles Match

for i in range(iterations):
    while (teamAScore < gameEnd and teamBScore < gameEnd) or winby2 < 2:
            
        if random.random() <= probA:
            
            probA = probAServe
            
            if servingTeam == 1:
                teamAScore +=1
                winby2 = abs(teamAScore - teamBScore)
#                 print(f"A {teamAScore}")
#                 print(f"B {teamBScore}")
            else:
                servingTeam = 1
#                 print("side out to A")

        else:
            probA = 1- probBServe
        
            if servingTeam == 2:
                teamBScore +=1
                winby2 = abs(teamAScore - teamBScore)
#                 print(f"A {teamAScore}")
#                 print(f"B {teamBScore}")
            else:
                servingTeam = 2
#                 print("side out to B")



                
    if teamAScore > teamBScore:
        teamAWins +=1
    
    else:
        teamBWins +=1
        
        
        
    
#     if teamAScore > teamBScore:
#         print ("Game Over, Team A Wins")
#     else:
#         print ("Game Over, Team B Wins")
        
    
    teamAScore = 0

    teamBScore = 0

    winby2 = abs(teamAScore - teamBScore)

    servingTeam = 1
    
    probA = probAServe
    
    
    
        
print(f"A Total Wins {teamAWins}")
print(f"B Total Wins {teamBWins}")
```

```python
gameEnd = 11

probAServe = .35

probBServe = .35

probA = probAServe


iterations = 100000

serveNum = 2


teamAScore = 0

teamBScore = 0

winby2 = abs(teamAScore - teamBScore)

servingTeam = 1

teamAWins = 0

teamBWins = 0
```

```python
#Doubles Match

for i in range(iterations):
    while (teamAScore < gameEnd and teamBScore < gameEnd) or winby2 < 2:

                 
        if random.random() <= probA:

            probA = probAServe

            if servingTeam == 1:
                teamAScore +=1
                winby2 = abs(teamAScore - teamBScore)
#                     print(f"A {teamAScore}  B {teamBScore}")

            else:
                if serveNum == 1:
                    serveNum += 1
#                         print("A Wins point, Second Serve to B")

                else:
                    serveNum = 1
                    servingTeam = 1
#                         print("Side out to A")

        else:
            probA = 1- probBServe

            if servingTeam == 2:
                teamBScore +=1
                winby2 = abs(teamAScore - teamBScore)
#                     print(f"A {teamAScore}  B {teamBScore}")

            else:
                if serveNum == 1:
                    serveNum += 1
#                         print("B Wins point, Second Serve to A")

                else:
                    serveNum = 1
                    servingTeam = 2
#                         print("Side out to B")



                
    if teamAScore > teamBScore:
        teamAWins +=1
#         print ("Game Over, Team A Wins")
    
    else:
        teamBWins +=1
#         print ("Game Over, Team B Wins")

        

    
    
    
    teamAScore = 0

    teamBScore = 0

    winby2 = abs(teamAScore - teamBScore)

    servingTeam = 1
    
    probA = probAServe
    
    serveNum = 2
    
    
    
        
print(f"A Total Wins {teamAWins}")
print(f"B Total Wins {teamBWins}")
```
